```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df=pd.read_csv("Highest Holywood Grossing Movies.csv")
df.head()
```

```python
plt.figure(figsize=(18,18))
sns.barplot(data=df,x='Distributor',y='World Sales (in $)')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(18,10))
df.groupby('Distributor')['World Sales (in $)'].sum().nlargest(10).plot(kind='barh',color='gold')
```
