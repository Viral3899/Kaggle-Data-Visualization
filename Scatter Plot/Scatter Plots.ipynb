```python
#Importing Dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=(16,8)
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
#Reading Dataset
df=pd.read_csv("insurance.csv")
df.head()
```

```python
# Scatter Plot Between BMI & CHARGES
plt.figure(figsize=(16,8))
sns.scatterplot(data=df,x='bmi',y='charges');
```

```python
# for Checking Correlation or Regline Plot Regplot
plt.figure(figsize=(16,8))
sns.regplot(data=df,x='bmi',y='charges',)
```

```python
# Scatter Plot With Seperating Smoker YES or No
plt.figure(figsize=(16,8))
sns.scatterplot(data=df,x='bmi',y='charges',hue='smoker')
```

```python
#Lmplot for BMI,SMOKER & CHARGES
sns.lmplot(data=df,x='bmi',y='charges',hue='smoker',height=8)
```

```python
#Swarmplot for SMOKER & CHARGES
plt.figure(figsize=(16,8))
sns.swarmplot(data=df,x='smoker',y='charges');
```
