```python
#importing Dependancies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
#Reading Dataset
df=pd.read_csv("iris.csv")
df
```

```python
#plotting Histogram for Petal Length
plt.figure(figsize=(16,8))
sns.histplot(data=df,x='Petal Length (cm)',kde=True,color='navy')
```

```python
#KDE plot
plt.figure(figsize=(16,8))
sns.kdeplot(data=df,x='Petal Length (cm)',shade=True,color='g')
```

```python
# 2D KDE plot
sns.jointplot(data=df,x='Petal Length (cm)', y='Sepal Width (cm)', kind="kde",height=8,color='firebrick')
```

```python
df_setosa=pd.read_csv("iris_setosa.csv")
df_versicolor=pd.read_csv("iris_versicolor.csv")
df_virginica=pd.read_csv("iris_virginica.csv")
```

```python
# Histograms for each species

plt.figure(figsize=(16,10))

sns.distplot(x=df_setosa['Petal Length (cm)'],label='Setosa',kde=False,color='chocolate')

sns.distplot(x=df_versicolor['Petal Length (cm)'],label='versicolor',kde=False,color='lawngreen')

sns.distplot(x=df_virginica['Petal Length (cm)'],label='virginica',kde=False,color='deeppink')

plt.legend(ncol=3)
```

```python
# KDE plots for each species
plt.figure(figsize=(16,10))

sns.kdeplot(x=df_setosa['Petal Length (cm)'],label='Setosa',color='chocolate',shade=True)

sns.kdeplot(x=df_versicolor['Petal Length (cm)'],label='versicolor',color='lawngreen',shade=True)

sns.kdeplot(x=df_virginica['Petal Length (cm)'],label='virginica',color='deeppink',shade=True)

plt.legend(ncol=3)
```
