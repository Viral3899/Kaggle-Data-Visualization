```python
#Importing Dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Reading Dataset
df=pd.read_csv("spotify.csv",index_col='Date',parse_dates=['Date'])
df.head()
```

```python
# Lineplot of Daily Global Streams of Each Songs

# Set the width and height of the figure
plt.figure(figsize=(16,8))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song 
sns.lineplot(data=df,palette='rainbow')
plt.xlabel("Date")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,8))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=df['Shape of You'], label="Shape of You",color='mediumspringgreen')

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=df['Despacito'], label="Despacito",color='gold')

# Add label for horizontal axis
plt.xlabel("Date")
```

```python
# Line chart showing daily global streams of 'Shape of You'
plt.figure(figsize=(16,8))
sns.lineplot(data=df['Shape of You'], label="Shape of You",color='violet')
```
